```r
Loaddata(file=system.file("extdata", "data_format.csv", package = "lilikoi"))
```

```r
# Metabolomics data matrix to pathway matrix

Metabolite_pathway_table=MetaTOpathway('name')
head(Metabolite_pathway_table)
```

```r
# Using PDSfun function to create Pathway matrix
PDSmatrix= PDSfun(Metabolite_pathway_table)
```

```r
# Show Pathway matrix
head(PDSmatrix)
dim(PDSmatrix)
```

```r
#Feature Selection

selected_Pathways_Weka= featuresSelection(PDSmatrix,threshold=0.1,method="info")
selected_Pathways_Weka
```

```r
#Testing Machine learning models of Lilikoi
machine_learning(PDSmatrix,selected_Pathways_Weka);
```
